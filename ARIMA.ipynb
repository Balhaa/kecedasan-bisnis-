# Contoh Data Sembarang: data penjualan selama 24 bulan secara random

```python
import pandas as pd
import numpy as np

np.random.seed(42)
data = np.random.poisson(lam=200, size=24)
df = pd.Series(data, index=pd.date_range(start='2023-01-01', periods=24, freq='M'))
```

# Model ARIMA: Gunakan ARIMA dari statsmodels

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df, order=(1,1,1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=6)
```

Plot Hasilnya:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df, label='Data Historis')
plt.plot(forecast.index, forecast, label='Forecast', color='red')
plt.legend()
plt.title('Forecast Penjualan 6 Bulan ke Depan')
plt.show()
```
